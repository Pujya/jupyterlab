```python
class BankAccount:
    def __init__(self):
        self.balance=0
    def deposit(self,amount):
        self.balance+=amount
        #a+=b Means simply a=a+b
    def withdraw(self,amount):
        if self.balance>=amount:
            self.balance-=amount
        else:
            print('Insufficient Balance')
    def check_balance(self):
        print('Current Balance is: ', self.balance)
```

```python
ob1=BankAccount()
ob1.deposit(2000)
ob1.check_balance()
ob1.withdraw(200)
ob1.check_balance()
ob1.withdraw(2000)
```

```python
class MinimumBalanceAccount(BankAccount):
    min_balance=5000
    def withdraw(self,amount):
        if amount>self.min_balance:
            amount=amount-self.min_balance
            print("Minimum Balance: ",self.min_balance)
            print("Can Only Withdraw: ", amount)
ob2=MinimumBalanceAccount()
ob2.withdraw(7000)
```

```python
# Import matplotlib library
import matplotlib.pyplot as plt
# Import data 
data=pd.read_csv('customer_churn.csv')
data.head()
```

```python
# Dividing the data in key value pair
key=data['Contract'].value_counts().keys().tolist()
# Extracting all values
value=data['Contract'].value_counts().tolist()
# Bar plot for contract column 
plt.bar(key,value,color="orange")
# Assign names to x, y axis
plt.xlabel('contract type of the customers')
plt.ylabel('count')
plt.title('distribution of the contracts')
```

```python
# Build Histogram for Monthly Charges
# We do not have to get key value for Monthly charges column because all values are in float and not object datatype
plt.hist(data['MonthlyCharges'], color='forestgreen', bins=20)
# Assign x and Y labels
plt.xlabel('monthly charges incurred')
plt.ylabel('count')
plt.title('distribution of the monthly charges')
```

```python
# Build Scatter Plot
# dropna() function helps drop any na values in data for the column listed
x=data['tenure'].dropna().head(10)
y=data['TotalCharges'].dropna().head(10)
plt.scatter(x,y, color='indigo')
plt.xlabel('Tenure of the customer')
plt.ylabel('Total charges incurred')
plt.title('Total charges vs Tenure')
```

```python
# Box plot between MonthlyCharges and PaymentMethod
# Take another library called seaborn
import seaborn as sns
datas=data
sns.boxplot(x='PaymentMethod',y='MonthlyCharges', data=datas, width=0.8)
plt.xlabel('Payment Method of cutomer')
plt.ylabel('Monthly Charges Incurred')
plt.title('Monthly Charges vs Payment Method')
```
